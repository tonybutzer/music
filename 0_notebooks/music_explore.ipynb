```python
from pathlib import Path

my_songs = []
for path in Path(dir_path).rglob('*.mp3'):
    s_dict={}
    s_dict['filepath']=str(path)
    try:
        i = eyed3.load(path)
        s_dict['artist'] = i.tag.artist
        s_dict['title'] = i.tag.title
        my_songs.append(s_dict)
    except:
        print(path)
```

```python
my_songs
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(my_songs)
```

```python
import pickle
# open a file, where you ant to store the data
file = open('songs_df.pickle', 'wb')

# dump information to that file
pickle.dump(df, file)

# close the file
file.close()
```
