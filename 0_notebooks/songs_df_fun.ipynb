```python
a = df.artist.unique()
```

```python
s_dfs=[]
cnt=0
for i in a:
    if str(i).startswith('A'):
        print(i)
        s_df = df[df['artist'].str.contains(str(i), na=False)]
        s_dfs.append(s_df)
        cnt=cnt+1
```

```python
import pandas as pd
df_c = pd.concat(s_dfs)
```
